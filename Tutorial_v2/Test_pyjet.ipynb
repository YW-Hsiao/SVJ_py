#### Note: Check which they use transformations.
In Particle Physics, we construct a new parameter which called rapidity as
 \begin{align}
 y = \frac{1}{2}\log{\left(\frac{E+p_z}{E-p_z}\right)}
 \end{align}
The transformation relations, with the $z$-axis as the beam axis, are
 \begin{align}
 p_x &= p_T\cos{\phi} \\
 p_y &= p_T\sin{\phi} \\
 p_z &= m_T\sinh{y} \\
 E &= m_T\cosh{y}
 \end{align}
where transverse mass $m_T$ and transverse momentum $p_T$ are given by
 \begin{align}
 m_T^2 &= m^2+p_x^2+p_y^2 \\
 p_T &= p\sin{\theta}
 \end{align}
In detectors, we take $p\gg m$, the rapidity becomes pseudorapidity is used:
 \begin{align}
 \eta = -\ln\left(\tan{\frac{\theta}{2}}\right)\approx y
 \end{align}
with values between $(-\infty,\infty)$. Thus, linear momentum is
 \begin{align}
 \vec{p} = (p_T,\eta,\phi)
 \end{align}
And we have easier transformation relation
 \begin{align}
 p_z &\approx m_T\sinh{\eta} \approx p_T\sinh{\eta} = p\sin{\theta}\cot{\theta} \\
 \left|\vec{p}\right| &= p_T\cosh{\eta} = p\sin{\theta}\csc{\theta}
 \end{align}

```python
print('This code is transform ATLAS coordinate to Cartesian coordinate.')

print('\n{:^14}{:^14}{:^14}{:^14}{:^14}{:^14}'.format('E(mt)', 'E(pt)', 'px', 'py', 'pz(mt)', 'pz(pt)'))
print('-'*84)
for constit in jets[5]:
#     print(constit.pt, constit.eta, constit.phi, constit.mass, constit.id)
    pt = constit.pt
    eta = constit.eta
    phi = constit.phi
    m = constit.mass
    
    mt = m**2 + pt**2
    px = pt*np.cos(phi)
    py = pt*np.sin(phi)
#     pz = mt * np.sinh(y)  # Not define y, since y is defined by E and pz simultaneously.
    pz_mt = mt*np.sinh(eta)
    pz_pt = pt*np.sinh(eta)
#     E = mt**2 * np.cosh(y)  # Not define y, since y is defined by E and pz simultaneously.
    E_mt = np.sqrt(m**2 + pt**2 + pz_mt**2)
    E_pt = np.sqrt(m**2 + pt**2 + pz_pt**2)
    print('{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}'.format(E_mt, E_pt, px, py, pz_mt, pz_pt))
    
    
print('\nor (E, px, py, pz):')
print(jets[5].constituents_array(ep=True))


print('\nConclusion: pyjet use E^2=m^2+pT^2+pz^2, px=pTcos(phi), py=pTsin(phi), pz=pTsinh(eta)!!')
print('='*42+'Check'+'='*42)
print('{:^14}{:^14}{:^14}{:^14}{:^14}{:^14}'.format('E(mt)', 'E(pt)', 'px', 'py', 'pz(mt)', 'pz(pt)'))
print('-'*84)
for i, constit in enumerate(jets[5]):
    pt = constit.pt
    eta = constit.eta
    phi = constit.phi
    m = constit.mass
    
    mt = m**2 + pt**2
    px = pt*np.cos(phi)
    py = pt*np.sin(phi)
    pz_mt = mt*np.sinh(eta)
    pz_pt = pt*np.sinh(eta)
    E_mt = np.sqrt(m**2 + pt**2 + pz_mt**2)
    E_pt = np.sqrt(m**2 + pt**2 + pz_pt**2)
    pyjet_Exyz=jets[5].constituents_array(ep=True)[i]
    diff_E_mt = E_mt-pyjet_Exyz[0]
    diff_E_pt = E_pt-pyjet_Exyz[0]
    diff_px = px-pyjet_Exyz[1]
    diff_py = py-pyjet_Exyz[2]
    diff_pz_mt = pz_mt-pyjet_Exyz[3]
    diff_pz_pt = pz_pt-pyjet_Exyz[3]
    
    print('{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}{:^14.8f}'.format(diff_E_mt*100/pyjet_Exyz[0], diff_E_pt, diff_px,
                                                                          diff_py, diff_pz_mt*100/pyjet_Exyz[3], diff_pz_pt))
```
